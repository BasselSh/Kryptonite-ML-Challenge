```python
def show_points(path, folder, name, df):
    img = Image.open(os.path.join(path, folder, name))
    coord = np.array(df.loc[(df['folder'] == int(folder)) & (df['image'] == name)])[0][2::]
    plt.figure(figsize = (6, 6))
    plt.imshow(img)
    plt.scatter(coord[::2], coord[1::2], s = 50, marker ='.', c='w')
    plt.axis('off');
```

```python
def full_eye_poligon(path, folder, name, df):
    img = Image.open(os.path.join(path, folder, name))
    coord = np.array(df.loc[(df['folder'] == int(folder)) & (df['image'] == name)])[0][2::]
    plt.figure(figsize = (6, 6))
    fig, ax = plt.subplots()
    plt.imshow(img)
    x = 0
    y = coord[1] - 20 # you may change the height
    width = 224
    height = 40 # you may change the height
    rectangle = mpl.patches.Rectangle((x, y), width, height, linewidth=2, facecolor='black', lw=2)
    ax.add_patch(rectangle)
    plt.axis('off');
```
